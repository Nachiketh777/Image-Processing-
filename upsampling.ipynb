```python
import cv2
image=cv2.imread('33.jpg')
print('size of image before pyrUp:',image.shape)
```

```python
image=cv2.pyrUp(image)
print('size of image after pyrUp:',image.shape)
cv2.imshow('Upsample',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.pyrDown(image)
print("Size of image after pyrDown: ", image.shape)
cv2.imshow('DownSample', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.image as img
import numpy as np
m=img.imread("taj.png")
w,h=m.shape[:2]
xNew=int(w*1/2);
yNew=int(h*1/2);
xScale = xNew/(w-1);
yScale = yNew/(h-1);
newImage = np.zeros([xNew, yNew, 4]);
for i in range(xNew-1):
    for j in range(yNew-1):
        newImage[i + 1, j + 1]= m[1 + int(i / xScale),1 + int(j / yScale)]
img.imsave('scaled.png', newImage);
```
